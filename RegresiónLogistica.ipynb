```python
data[['age','currentSmoker']].plot.scatter(x='age',y='currentSmoker')
```

```python
# pruebas de parametro
w = 0.09
b = -3.6
```

```python
# despues de hacer el modelos (se explica más adelante)
# intercepto (b): [-3.68596089]
# pendiente (w): [[0.09351691]]
```

```python
# puntos de la recta
x = np.linspace(0,data['age'].max(),100)
y = 1/(1+np.exp(-(w*x+b)))

# grafica de la recta
data.plot.scatter(x='age',y='currentSmoker')
plt.plot(x, y, '-r')
plt.ylim(0,data['currentSmoker'].max()*1.1)
# plt.grid()
plt.show()
```
